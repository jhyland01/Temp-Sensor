```python
from autogluon.tabular import TabularPredictor

# Load the saved predictor
saved_path = '/Users/johnny/Library/CloudStorage/OneDrive-Personal/py/RES_hackathons/Temp Sensor/AutogluonModels/ag-20230724_154150'  # Replace with the path where your model is saved
predictor = TabularPredictor.load(saved_path)

# Get and print the leaderboard
leaderboard = predictor.leaderboard()
print(leaderboard)

# Fetch detailed information including hyperparameters
model_details = predictor.get_model_attribute_dict()

# Loop to get hyperparameters of each model
for model_name in model_details:
    print(f"Hyperparameters for model {model_name}:")
    print(model_details[model_name]['hyperparameters'])
```

```python
# Access the internal bagged learner model
learner = predictor._learner

# Get model names
model_names = learner.get_info()

# Loop through each model to extract hyperparameters
for model_name in model_names:
    model = learner.load_trained_model(model_name)
    print(f"Model Name: {model_name}")
    print(f"Hyperparameters: {model.params}")
```
